```python
#importieren nur einen Teil der urllib Library, pd ist community standard abkürzung für pandas, ist wunderbar dokumentiert und haben das alles schon über Anakonca mitbekommen
import urllib.request
import pandas as pd
```

```python
#haben hier einfach nur einen String, die URL hinterlegt, aber noch nichts gemacht Aktualisieren kann man dann mit noch einmal laufen lassen der Zelle 
open_apc_url = "https://raw.githubusercontent.com/OpenAPC/openapc-de/master/data/fuberlin/APC_FU_Berlin_2015.csv"
```

```python
apc_file = "openapc.csv"
```

```python
#jetzt wollen wir automatisiert die Daten holen, um weniger händisch zu machen, immer aktuell zu sein(muss man aber erst noch angeben, momentan passiert es nicht, außer wenn ich die Zeile für den Download neu lade), reproduzierbar, die Funktion bruacht Quelle und Ziel
#wenn man ganz viele Dateien runterladen will, könnte man auch einen Vorloop machen
urllib.request.urlretrieve(open_apc_url, apc_file)
```

```python
apcs = pd.read_csv(apc_file)
```
